```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
chr = pd.read_csv("data/County Health Rankings/2023_CHR_Analytic_Data.csv")
```

### Get relevant rows (state-level data)

List of States

```python
states = [
    "Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado",
    "Connecticut", "Delaware", "District of Columbia", "Florida", "Georgia",
    "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky",
    "Louisiana", "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota",
    "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada", "New Hampshire",
    "New Jersey", "New Mexico", "New York", "North Carolina", "North Dakota",
    "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island",
    "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Vermont",
    "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming"
]
```

Filter and keep only state-level measures

```python
states_data = chr[chr["Name"].isin(states)]
states_data
```

```python
states_data.shape
```

Remove duplicated row for DC

```python
states_data.drop(330, inplace= True)
```

### Get relevant columns

```python
cols = [
    "Name",
    "Air Pollution - Particulate Matter raw value",
    "Drinking Water Violations raw value", 
    "Food Environment Index raw value",
    "Access to Exercise Opportunities raw value",
    "Limited Access to Healthy Foods raw value"
]

filtered_data = states_data[cols]
filtered_data
```

Convert column names

```python
filtered_data.rename(columns={
    "Name": "state",
    "Air Pollution - Particulate Matter raw value": "avg_pm25",
    "Drinking Water Violations raw value": "water_violation_rate",
    "Food Environment Index raw value": "food_env_index",
    "Access to Exercise Opportunities raw value": "exercise_access",
    "Limited Access to Healthy Foods raw value": "limited_healthyfood_access"
}, inplace=True)
```

```python
filtered_data
```

Convert percentages to more readable numbers

```python
def to_percentage(x):
    percentage = float(x) * 100
    rounded = round(percentage, 2)
    return rounded
```

```python
filtered_data["water_violation_rate"] = filtered_data["water_violation_rate"].apply(to_percentage)
filtered_data["exercise_access"] = filtered_data["exercise_access"].apply(to_percentage)
filtered_data["limited_healthyfood_access"] = filtered_data["limited_healthyfood_access"].apply(to_percentage)
```

```python
filtered_data
```

Save

```python
filtered_data = filtered_data.reset_index(drop=True)
```

```python
filtered_data.to_csv("processed_data/chr_processed.csv")
```
