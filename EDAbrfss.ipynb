###  Columns description

* _STATE: State FIPS Code
* SEXVAR: Sex of Respondent
* _AGEG5YR: Reported age in five-year age categories calculated variable
* EDUCA: Education Level
* INCOME3: Income Level (Annual household income from all sources)
* _RACE: Race-Ethnicity grouping
* BPHIGH6: Ever Told Blood Pressure High
* DIABETE4: (Ever told) you had diabetes
* CVDINFR4: (Ever told) you had a heart attack, also called a myocardial infarction?
* CVDCRHD4: (Ever told) (you had) angina or coronary heart disease?
* CVDSTRK3: Ever Diagnosed with a Stroke
* SMOKE100: Smoked at Least 100 Cigarettes in entire life
* SMOKDAY2: Frequency of Days Now Smoking
* EXERANY2: Adults who reported doing physical activity or exercise during the past 30 days other than their regular job
* _BMI5: Body Mass Index (BMI)
* _BMI5CAT: Four-categories of Body Mass Index (BMI)
* _RFBMI5: Adults who have a body mass index greater than 25.00 (Overweight or Obese)
* ALCDAY4: Alcohol consumption in past 30 days
* GENHLTH: Self-rated General Health
* PHYSHLTH Number of Days Physical Health was Not Good on last 30 days
* MENTHLTH: Number of Days Mental Health was Not Good on last 30 days
* CHOLCHK3: How Long since Cholesterol Checked
* CHECKUP1: Length of time since last routine checkup

### Check for missing data

Codes for missing values: Refused/Not asked/Missing

* BPHIGH6: 7 and 9
* DIABETE4: 7 and 9
* CVDINFR4: 7 and 9
* CVDCRHD4: 7 and 9
* CVDSTRK3: 7 and 9
* SMOKE100: 7 and 9
* SMOKDAY2: 7 and 9
* EXERANY2: 7 and 9
* _BMI5: BLANK
* _BMI5CAT: BLANK
* _RFBMI5: 9
* ALCDAY4: 777, 999
* GENHLTH: 7 and 9 
* PHYSHLTH: 77 and 99
* MENTHLTH: 77 and 99
* CHOLCHK3: 7 and 9
* CHECKUP1: 7 and 9

Create function to count missing data

```python
def check_missing_data(df, column, values):
    '''
    df: data
    column: column name (string) to check
    values: set of values considered as "missing"
    
    Returns proportion of total missing
    '''

    # Total rows
    total = len(df)

    # Boolean masks
    custom_missing_mask = df[column].isin(values)
    nan_mask = df[column].isna()

    # Counts
    custom_missing_count = custom_missing_mask.sum()
    nan_count = nan_mask.sum()
    total_missing = custom_missing_count + nan_count

    # Proportion
    missing_proportion = (total_missing / total) * 100

    return np.round(missing_proportion, 2)
```

Run for different cases and store results in dictionary

```python
missing_dict = {}

#7 and 9 as missing data
first_batch = [
    'BPHIGH6',
    'DIABETE4',
    'CVDINFR4',
    'CVDCRHD4',
    'CVDSTRK3',
    'SMOKE100',
    'SMOKDAY2',
    'EXERANY2',
    'GENHLTH',
    '_RFBMI5',
    'CHOLCHK3',
    'CHECKUP1'
]

for col in first_batch:
    result= check_missing_data(brfss_data, col, {7.0,9.0})
    missing_dict[col] = result

#Only BLANK/NaN values
second_batch = ["_BMI5", "_BMI5CAT"]

for col in second_batch:
    result= check_missing_data(brfss_data, col, {})
    missing_dict[col] = result

#77, 99, 777 and 999 as missing data
third_batch = ["ALCDAY4", "GENHLTH", "PHYSHLTH", "MENTHLTH"]

for col in third_batch:
    result= check_missing_data(brfss_data, col, {77, 99, 777, 999})
    missing_dict[col] = result
```

Plot

```python
plt.figure(figsize=(10, 6))
plt.bar(missing_dict.keys(), [v for v in missing_dict.values()], color = "skyblue")
plt.xlabel("Variable")
plt.ylabel("Proportion of missing data (%)")
plt.title("Missing data on health indicators")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
